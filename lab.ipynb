## Error Assigning

```python
import re
```

```python
with open('error.log', 'r') as f:
    error_msgs = f.readlines()
```

```python
def error_gen(error_msgs):
    reg = re.compile('\[(\w+)\] ([\w\d |[{:\']+)')
    for error_msg in error_msgs:
        res = reg.search(error_msg)
        author = res.group(1)
        error = res.group(2)
        yield (author, error)
```

```python
for author, error in error_gen(error_msgs):
    if authors[author] == -1:
        print(error)
```
